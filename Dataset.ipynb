```python
import numpy as np
import pandas as pd
```

```python
df_cust_data = pd.read_excel('mazda.xlsx','Customer data')
df_campaign_customers = pd.read_excel('mazda.xlsx','Campaigns_customers')
df_prospects_data = pd.read_excel('mazda.xlsx','Prospects data')
df_campaign_prospects = pd.read_excel('mazda.xlsx','Campaigns_prospects')
```

```python
def domainsplit(x):
    try:
        return x.split('@')[1].split('.')[0].lower()
    except:
        return None
```

```python
def business_email(x):
    if x is None:
        return None
    elif x in ['skynet', 'telenet', 'hotmail', 'msn', 'voo', 'gmail', 'yahoo', 'gamail', 'msn', 'ymail']:
        return "Private Email"
    else:
        return "Business Email"
```

```python
df_cust_data['domain_email'] = df_cust_data['Email Address'].apply(lambda x: domainsplit(x))
```

```python
df_cust_data['type_email'] = df_cust_data['domain_email'].apply(lambda x: business_email(x))
```
